```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = [
    "Linear Regression",
    "Polynomial\nRegression (Deg 2)",
    "Decision Tree",
    "Random Forest",
    "SVM",
    "Neural Network\n(Wide)"
]

test_rmse = [14530.87, 15503.13, 7735.98, 5696.41, 12063.52, 7672.10]
adj_r2 = [0.5078, 0.2344, 0.7681, 0.8743, 0.6608, 0.7524]

x = np.arange(len(models))
width = 0.35

fig, ax1 = plt.subplots(figsize=(12, 6))

# First bar: RMSE
bar1 = ax1.bar(x - width/2, test_rmse, width, label='Test RMSE', color='#f4a582')
ax1.set_ylabel('Test RMSE', color='#f4a582')
ax1.tick_params(axis='y', labelcolor='#f4a582')

# Twin axis for Adjusted R
ax2 = ax1.twinx()
bar2 = ax2.bar(x + width/2, adj_r2, width, label='Adjusted R', color='#92c5de')
ax2.set_ylabel('Adjusted R', color='#92c5de')
ax2.tick_params(axis='y', labelcolor='#92c5de')

# X-axis
ax1.set_xticks(x)
ax1.set_xticklabels(models, rotation=15)
ax1.set_title('Model Comparison: Test RMSE vs Adjusted R')

# Annotate values
def annotate_bars(ax, bars, decimals):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.{decimals}f}',
                    xy=(bar.get_x() + bar.get_width()/2, height),
                    xytext=(0, 3),
                    textcoords='offset points',
                    ha='center', va='bottom', fontsize=9)

annotate_bars(ax1, bar1, 0)
annotate_bars(ax2, bar2, 4)

# Legends
fig.legend(loc='upper right', bbox_to_anchor=(1, 1), bbox_transform=ax1.transAxes)

plt.tight_layout()
plt.show()
```
